```python
from PIL import Image
from scipy.linalg import svd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_image_feature(s, k):
    # 对s保留前k个特征值
    s_temp = np.zeros(s.shape[0])
    s_temp[0:k] = s[0:k]
    s = s_temp * np.identity(s.shape[0])
    # 用新的s及p，q重构矩阵
    temp = np.dot(p, s)
    temp = np.dot(temp, q)
    plt.imshow(temp, cmap=plt.cm.gray, interpolation='nearest')
    plt.show()
    print(A-temp)
```

```python
# 加载图片并将图片灰度化后转化为矩阵形式
image = Image.open('./husky.jpg').convert('L')
A = np.array(image)
# 原图像灰度化效果
plt.imshow(A, cmap=plt.cm.gray, interpolation='nearest')
plt.show
```

```python
# 对矩阵进行奇异值分解得p,s,q
p, s, q = svd(A, full_matrices=False)
# 愿图像特征值数量
n = A.shape[1]
# 奇异值数量为原有的1%
get_image_feature(s, int(0.01*n))
```

```python
# 奇异值数量为原有的10%
get_image_feature(s, int(0.1*n))
```
